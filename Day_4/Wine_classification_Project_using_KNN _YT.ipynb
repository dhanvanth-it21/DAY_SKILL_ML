```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
wine = datasets.load_wine()
```

```python
wine
```

```python
df = pd.DataFrame(wine["data"],columns  = wine["feature_names"])
df
```

```python
df["target"] = wine["target"]
df
```

```python
df.shape
```

```python
# to find the missing or Nan data in  the wine dataset
df.isna().sum()
```

## test train Split

```python
from sklearn.model_selection import  train_test_split
```

```python
X = df.drop(["target"],axis = "columns")
```

```python
y  = df["target"]
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test  = train_test_split(X,y,test_size = 0.2,random_state = 55)
```

## train classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

## tunnig sensitivity of the model to n_neighbors

```python
k_range = range(1,25)
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train,y_train)
    scores.append(knn.score(X_test,y_test))
    
plt.figure()
plt.xlabel(" k count")
plt.ylabel("model_accuracy")
plt.scatter(k_range, scores)
plt.grid()
plt.xticks([0, 5, 10, 20, 30])
plt.show()
```

```python
#  in the above graph we can see that point for every frequency sensitivity
# 3 has the high so to get better result we need to change the n_negihbors as 3
```

```python
test_sizes = [0.8,0.7,0.6,0.5,0.4,0.3,0.2,0.1]
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
plt.figure()
for test_size in test_sizes:
    scores = []
    
    for i in range(1,1000):
        X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 1-test_size)
        knn.fit(X_train,y_train)
        scores.append(knn.score(X_test,y_test))
    plt.plot(test_size,np.mean(scores), "bo")
plt.xlabel("Training_split %")
plt.ylabel("Acurracy")
plt.show()
```

## make prediction

```python
prediction = knn.predict(X_test)
prediction
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,prediction)
cm
```

```python
import seaborn as sns
plt.figure(figsize = (8,7))
sns.heatmap(cm,annot= True)
plt.title("Confusion matrix")
plt.xlabel("Truth")
plt.ylabel("Prediction")
plt.show()
```
